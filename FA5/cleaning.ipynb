```python
import pandas as pd
import datetime
```

```python
df = pd.read_excel("phivolcs_earthquakes_2018_2025.xlsx")
```

```python
df.head(5)
```

### Cleaning

```python

df_clean = df.copy()
df_clean.head(5)
```

```python

df_clean['Datetime'] = df_clean['Date - Time (Philippine Time)'].astype(str)
df_clean['Datetime'] = df_clean['Datetime'].str.replace(r'_x000D_|\\n|\\t', ' ', regex=True)
df_clean['Datetime'] = df_clean['Datetime'].str.replace(r'\\s+', ' ', regex=True).str.strip()


df_clean['Datetime'] = pd.to_datetime(df_clean['Datetime'], errors='coerce', dayfirst=True)
```

#### Location

```python
df_clean['Location'] = df_clean['Location'].str.replace(r'_x000D_\\n\\t\\t', ' ', regex=True)
df_clean['Location'] = df_clean['Location'].str.replace(r'_x000D_|\n|\t', ' ', regex=True)
df_clean['Location'] = df_clean['Location'].str.replace(r'\s+', ' ', regex=True).str.strip()

# df_clean['Province'] = df_clean['Location'].str.extract(r'\(([^)]+)\)', expand=False)
```

```python
df_clean['Location'] = df_clean['Location'].str.replace(r'^.*\bof\b\s+', '', regex=True)
df_clean['City'] = df_clean['Location'].str.replace(r'\(([^)]+)\)', '', regex=True)
df_clean['Province'] = df_clean['Location'].str.extract(r'\(([^)]+)\)', expand=False)

df_clean.head(5)  
```

```python

df_clean = df_clean.rename(columns={'Latitude (ºN)': 'Lat', 'Longitude (ºE)': 'Long', 'Depth (km)': 'Depth_km'})
```

```python
df_clean = df_clean[['Datetime', 'Lat', 'Long', 'Depth_km', 'Mag', 'City','Province']]
df_clean.head(5)
```

```python
null = df_clean[df_clean.isnull().any(axis=1)]
null.head(5)
null['City'].unique()
```
