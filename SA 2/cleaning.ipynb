```python
df_clean = df
df_clean['Trip Start Timestamp'] = pd.to_datetime(df['Trip Start Timestamp'], format="%m/%d/%Y %I:%M:%S %p")
df_clean['Date'] = df_clean['Trip Start Timestamp'].dt.date
df_clean['Time'] = df_clean['Trip Start Timestamp'].dt.time
df_clean.head(3)
```

```python
df_clean.head(1)
```

```python
df_clean['Month'] = df_clean['Trip Start Timestamp'].dt.strftime('%B')
df_clean['Year'] = df_clean['Trip Start Timestamp'].dt.year
df_clean.head(1)
```

```python
# monthly_counts = df_clean.groupby(['Month', 'Year','Date', 'Pickup Community Area']).size().reset_index(name='Trip Count')
```

```python
for col in ['Trip Total', 'Fare', 'Trip Miles']:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
```

```python
monthly_agg = df_clean.groupby([
    'Year',
    'Month',
    'Pickup Community Area',
    'Dropoff Community Area',
    'Company',
    'Payment Type'
]).agg({
    'Trip ID': 'count',
    'Trip Total': 'mean',
    'Fare': 'mean',
    'Tips': 'mean',
    'Trip Miles': 'mean'
}).reset_index()

monthly_agg.head(5)
```

```python
monthly_agg['Date'] =monthly_agg['Month'] + '' + monthly_agg['Year'].astype(str)
monthly_agg.to_csv('monthly_aggregated_taxi_fares.csv', index=False)
```
