```python
import pandas as pd
```

```python
df = pd.read_csv('senate25-final_updated.csv')
df.head(5)
```

```python
# df.columns
# df.info()
df.columns[12:]
```

```python
df.isna().sum()
```

```python
candidates = df.columns[12:]

votes = df[candidates].sum().reset_index()
votes.columns = ['candidate', 'votes']
votes.sort_values(by='votes', ascending=False)
```

```python
grouped = df.groupby('municipality')[candidates].sum().reset_index()


votes_municipality =grouped.melt(id_vars='municipality', var_name='Candidate', value_name='Votes')
votes_municipality
```

```python
melted = df.melt(
    id_vars=['region', 'province', 'municipality', 'barangay'],
    value_vars=candidates,
    var_name='Candidate',
    value_name='Votes'
)

melted['Candidate'] = melted['Candidate'].str.replace(r'^\d+\.\s*', '', regex=True)



# votes_by_province.to_csv("votes_per_candidate_province.csv", index=False)
```

```python
votes_by_province = melted.groupby(['province', 'Candidate'], as_index=False)['Votes'].sum()
votes_by_province
```

```python
votes_by_region = melted.groupby(['region', 'Candidate'], as_index=False)['Votes'].sum()
votes_by_region
```

```python
df_grouped = df.groupby(['region', 'province', 'municipality'], as_index=False).agg({
    'registeredVoters': 'sum',
    'actualVoters': 'sum',
    'overVotes': 'sum',
    'underVotes': 'sum'
})

df_grouped
```
