```python
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)


years = list(range(2018, 2026))
months = [
    "January", "February", "March", "April", "May", "June", "July",
    "August", "September", "October", "November", "December"
]


headers = [
    "Date - Time (Philippine Time)",
    "Latitude (ºN)",
    "Longitude (ºE)",
    "Depth (km)",
    "Mag",
    "Location"
]

all_data = []

for year in years:
    for i, month in enumerate(months):
        if year == 2025 and i > 6:
            break  # stop at July 2025

        url = f"https://earthquake.phivolcs.dost.gov.ph/EQLatest-Monthly/{year}/{year}_{month}.html"
        print(f"Fetching: {url}")

        
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}, timeout=10, verify=False)
        if response.status_code != 200:
            print(f"Page not found: {url}")
            continue

        soup = BeautifulSoup(response.content, "html.parser")
        tables = soup.find_all("table")

        if len(tables) < 3:
            print(f"Less than 3 tables found in {url}, skipping.")
            continue

        table = tables[2]
        rows = []

        for tr in table.find_all("tr")[1:]:
            cells = [td.text.strip() for td in tr.find_all("td")]
            if len(cells) == len(headers):
                rows.append(cells)

        if rows:
            df = pd.DataFrame(rows, columns=headers).drop(0)
            all_data.append(df)




```

```python
final_df = pd.concat(all_data, ignore_index=True)
final_df.tail(10)
```
